```python
import requests
```

```python
url = 'https://api2.binance.com/api/v3/ticker/24hr'
```

```python
response = requests.get(url)
```

```python
data = response.json()
```

```python
data
```

```python
symbols = []
for x in data:
    p=x.get("symbol")
    symbols.append(p)
symbols
```
