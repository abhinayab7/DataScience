## Build Chatbot

```python
from langchain_core.messages import HumanMessage,AIMessage,SystemMessage
user_input = input("Enter you message or type 'exit' to end the conversation")
messages=[SystemMessage(content="You are a helpful assistant")]
while user_input!='exit':
  messages.append(HumanMessage(content=user_input))
  response = llm.invoke(messages)
  print("Response from LLM: ",response.content)
  messages.append(AIMessage(content=response.content))
  print("Message list after adding LLM output: ",messages)
  user_input = input("Enter you message or type 'exit' to end the conversation")
```

```python
from langchain.tools.tavily_search import TavilySearchResults
from crewai.tools import BaseTool
from pydantic import BaseModel
from pydantic.fields import Field

class ToolInput(BaseModel):
    question: str = Field(description="search query")

class SearchTool(BaseTool):
  name: str = "search tool"
  description: str = "use this tool to find current information from web required for user query"
  args_schema: ToolInput

  def _run(self, question: str):
    "use this tool to find current information from web required for user query"
    return TavilySearchResults().run(question)
```

```python
from crewai.tools import BaseTool
from pydantic import BaseModel
from pydantic.fields import Field

class ToolInput(BaseModel):
    question: str = Field(description="The question to ask the LLM.")

class GenerationTool(BaseTool):
  name: str = "generation tool"
  description: str = "use this tool to generate content based on LLM own knowledge"
  args_schema: ToolInput

  def _run(self, question: str):
    "use this tool to generate content based on LLM own knowledge"
    return llm.run(question)
```

#Creating Agent

```python
from crewai import Agent
router_agent = Agent(
    role= "Router",
    goal= "route user question to tool",
    backstory=(
        "you are expert in analysing user question "
        "for a give user query, you need to suggest which tool we need to use to get the answer"
        "availble tools are geeration_tool , search_tool, pdf_tool"
    ),
    LLM=llm,
    verbose=True,
    allow_delegation=True
)
```

```python
from crewai import Task
Task(
    description=(
        "analyze context and keywords present in question : {question} "
        "based on context and keyword return a single word from list ['web_Search', 'generation', 'vectordb'] "
        "return word 'web_Search' if question need current iformation from internet"
    ),
    expected_output="A single word from the list ['web_Search', 'generation', 'vectordb'] based on the analysis of the question."
)
```
